Importing the sqlite3 and pandas libraries to get us started as well as initializing the 3 databases in memory.

```python
import sqlite3
import pandas as pd

# Load datasets
athletes = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/athlete_events.csv")
regions = pd.read_csv("https://raw.githubusercontent.com/rfordatascience/tidytuesday/master/data/2021/2021-07-27/noc_regions.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
sales = pd.read_csv("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv", sep='\t')


# Connect to SQLite in-memory DB
conn = sqlite3.connect(":memory:")

# Write DataFrames to SQL tables
athletes.to_sql("athletes_table", conn, index=False, if_exists="replace")
regions.to_sql("regions_table", conn, index=False, if_exists="replace")
sales.to_sql("sales_table", conn, index=False, if_exists="replace")
```

#Section 1
##Athelete's database
Total number of medals won by the top 5 countries

```python
pd.read_sql(
    """
    SELECT Count(medal) as total_medals, team
    FROM athletes_table
    GROUP BY team
    ORDER BY total_medals DESC
    LIMIT 5;
    """,conn
)
```

Average of atheletes who won a gold medal

```python
pd.read_sql(
    """
    SELECT AVG(age) as avg_age
    FROM athletes_table
    WHERE medal = 'Gold';
    """,conn
)
```

Number of distict events in each sport

```python
pd.read_sql(
    """
    SELECT sport, count(DISTINCT event) as number_of_events
    FROM athletes_table
    GROUP BY sport;
    """,conn
)
```

All athletes from United States

```python
pd.read_sql(
    """
    SELECT *
    FROM athletes_table
    WHERE NOC = 'USA';
    """,conn
)
```

Number of medals awarded each year

```python
pd.read_sql(
    """
    SELECT year, COUNT(medal) as number_of_medals
    FROM athletes_table
    GROUP BY year
    """,conn
)
```

Atheletes whose height or weight is missing

```python
pd.read_sql(
    """
    SELECT *
    FROM athletes_table
    WHERE height IS NULL or weight IS NULL
    """, conn
)
```

Replace missing height with average athlete height

```python
pd.read_sql(
    """
    SELECT *
    FROM athletes_table
    WHERE height = (
      SELECT COALESCE(height,AVG(height))
      FROM athletes_table
    )
    """,conn
)
```

##Sales dataset

Returning total sales value per item

```python
pd.read_sql(
    """
    SELECT item_name,
    SUM(CAST(REPLACE(item_price,'$','')AS REAL)) as total_sales_value
    FROM sales_table
    GROUP BY item_name
    """,conn
)
```

Show top 5 records with highest item_price

```python
pd.read_sql(
    """
    SELECT *
    FROM sales_table
    ORDER BY CAST(REPLACE(item_price,'$','')AS REAL) DESC
    LIMIT 5
    """,conn
)
```

Total number of unique customer orders(each order ID is a customer)

```python
pd.read_sql(
    """
    SELECT COUNT(DISTINCT order_id) as num_of_orders
    FROM sales_table
    """,conn
)
```

#Section 2

Find countries with high perfoming atheles in the olympics

```python
pd.read_sql(
    """
    WITH joined_table as (
    SELECT r.region, a.name, a.medal, a.age
    FROM regions_table r
    LEFT JOIN athletes_table a ON r.NOC = a.NOC
    )
    SELECT region, COUNT(DISTINCT name) as num_of_medalholders, AVG(age) as avg_age,
    CASE
      WHEN AVG(age) < 25 THEN 'high'
      WHEN AVG(age) >= 25 AND AVG(age) <= 30 THEN 'medium'
      ELSE 'low'
    END AS performance
    FROM joined_table
    WHERE medal IS NOT NULL
    GROUP BY region
    """,conn
)
```
